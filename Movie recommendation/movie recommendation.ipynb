```python
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L    
```

```python
#parsing only 3 main actor
def convert(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L    
```

```python
movies2['cast']=movies2['cast'].apply(convert)
```

```python
movies2.head()
```

```python
movies2['crew'][0]
```

```python
def fetch(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
        else:
            continue
    return L    
```

```python
movies2['crew']=movies2['crew'].apply(fetch)
```

```python
movies2['overview']=movies2['overview'].apply(lambda x:x.split())
```

```python
movies2['genres']=movies2['genres'].apply(lambda x:[i.replace(' ','') for i in x])
movies2['keywords']=movies2['keywords'].apply(lambda x:[i.replace(' ','') for i in x])
movies2['cast']=movies2['cast'].apply(lambda x:[i.replace(' ','') for i in x])
movies2['crew']=movies2['crew'].apply(lambda x:[i.replace(' ','') for i in x])
```

```python
 movies2['tags']=movies2['overview']+movies2['genres']+movies2['keywords']+movies2['cast']+movies2['crew']
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)    
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import  cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
list(similarity[0])
```

```python
#sorted(similarity[0],reverse=True)
#it will lose the index of the movie
#use enumarete to hold the index
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
#for i in x:
 #   print(i[0])
```

```python
def recommend(movie):
    movie_index=new_df[new_df['title']==movie].index[0]
    distances=similarity[movie_index]
    movie_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
   
    for i in movie_list:
        print(new_df.loc[i[0],'title'])
   
```
